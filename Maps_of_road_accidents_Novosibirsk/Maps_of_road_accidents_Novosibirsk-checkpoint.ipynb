```python
#создаем список в городе районов
# Калининский район состоит из 2 частей. Его добавим в конце методом исключения 
nsk_rayons = ['Дзержинский', 'Кировский', 'Ленинский', 'Октябрьский', 
              'Первомайский', 'Советский' , 'Центральный', 'Заельцовский', 'Железнодорожный']
```

Работаем с API open street maps 
- в цикле обращаемся к нему для каждого района из осзданного нами списка
- передавая параметр polygon_geojson=1 мы говорим, что нам нужны границы этого района
- выгружается строка, поэтому мы переводим ее в json - список словарей (у нас тут один словарь в списке)
- из этого json вытаскиваем координаты полигона - то есть границы района - это список списков
- переводим вытащенные координаты в полигон с помощью библиотеки shapely
- создаем таблицу с названиями районов и их границами

```python
lst_r = []
lst_poly = []
for r in nsk_rayons:
    d = json.loads(requests.get(f"http://nominatim.openstreetmap.org/search?q={r} район Новосибирск&polygon_geojson=1&format=json").text)[0]
    lst_sets = []
    for elem in d['geojson']['coordinates'][0]:
        lst_sets.append(tuple(elem))
    lst_r.append(r)
    lst_poly.append(Polygon(lst_sets))
```

```python
df_polygons = pd.DataFrame()
df_polygons['район'] = lst_r
df_polygons['geometry'] = lst_poly
df_polygons
```

Принадлежность точки к району.Для каждого полигона и точки (места аварии) проверяем вхождение точки в полигон

```python
list_rayons = []
list_points = []
list_polygons = []
list_id = []
for i in range(len(df_polygons)):
    for j in range(len(df)):
        if df_polygons['geometry'][i].contains(df['geometry'][j]):
            list_rayons.append(df_polygons['район'][i]) 
            list_points.append(df['geometry'][j])
            list_polygons.append(df_polygons['geometry'][i])
            list_id.append(df['id'][j])
```

```python
#Создадим колонку с сезонами 
# функция для категоризации
def categorize_season(season):
    try:
        if 3 <= season <= 5:
            return 'весна'
        elif 6 <= season <= 8:
            return 'лето'
        elif 9 <= season <= 11:
            return 'осень'
        else:
            return 'зима'
    except:
        pass

ready_data['season_category'] = ready_data['month'].apply(categorize_season)
```

```python
# Создадим колонки с полом ,стажем вождения и ролью в ДТП 
#Функция для years_of_driving_experience
def get_year(cell):
    try:
        return (json.loads(cell.replace("'", '"'))[0]['participants'][0]['years_of_driving_experience'])
    except: return np.nan
#Функция для role
def get_role(cell):
    try:
        return (json.loads(cell.replace("'", '"'))[0]['participants'][0]['role'])
    except: return np.nan
#Функция для gender
def get_gender(cell):
    try:
        return (json.loads(cell.replace("'", '"'))[0]['participants'][0]['gender'])
    except: return np.nan
```

```python
#Добавим новые колонки в df
ready_data['years_of_driving_experience'] = ready_data['vehicles'].apply(get_year)
ready_data['role'] = ready_data['vehicles'].apply(get_role)
ready_data['gender'] = ready_data['vehicles'].apply(get_gender)
```

- Данные представлены с `2015 по 2022` год. Если на `2015` год приходится `16%` ДТП, то на `2022` год - `11%`.

  - Самым проблемным месяцем является октябрь- `11%`
